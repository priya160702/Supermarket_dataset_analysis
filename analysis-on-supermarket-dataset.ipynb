```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop(['Invoice ID', 'Date', 'Time'], axis = 1)
```

```python
# cogs stands for cost of goods sold
df['gross margin percentage'].unique()
```

```python
df = df.drop(['gross margin percentage'], axis = 1)
```

## Visualizing the data

The below graph shows count of females and males is almost the same.

```python
print(df.shape)
df['Gender'].value_counts()
```

```python
sns.countplot(x='Gender', data=df)
```

## Trying to create dummies for gender column for getting numeric type data for seaborn bar plot

```python
gender_dummies  = pd.get_dummies(df['Gender'])
gender_dummies.head()
```

## Now, joining the original data frame (df) with gender dummies data frame (gender_dummies)

```python
df = pd.concat([df, gender_dummies], axis = 1)
df.head()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Product line', y = 'Female', data = df)
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Product line', y = 'Male', data = df)
```

## Plotting a heatmap to see if there is any correlation between columns

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
correlation_matrix = df_encoded.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
